```python
import turtle

#thiết lập màn hình và bút vẽ
sc=turtle.Screen()
sc.bgcolor("sky blue")
t=turtle.Turtle()
t.pensize(5)
# vẽ thân cây
t.penup()
t.goto(-100,-100)
t.left(90)
t.fillcolor("brown")
t.pendown()
t.begin_fill()
t.forward(100)
t.left(90)
t.forward(40)
t.left(80)
t.forward(100)
t.left(100)
t.forward(60)
t.end_fill()
# vẽ lá cây 1
t.penup()
t.goto(-120,0)
t.pendown()
t.fillcolor("green")
t.begin_fill()
t.forward(50)
t.left(120)
t.forward(100)
t.left(120)
t.forward(100)
t.left(120)
t.forward(50)
t.end_fill()
# vẽ lá cây 2
t.penup()
t.goto(-120,75)
t.pendown()
t.begin_fill()
t.forward(50)
t.left(120)
t.forward(100)
t.left(120)
t.forward(100)
t.left(120)
t.forward(50)
t.end_fill()
#vẽ lá cây 3
t.penup()
t.goto(-120,150)
t.pendown()
t.begin_fill()
t.forward(50)
t.left(120)
t.forward(100)
t.left(120)
t.forward(100)
t.left(120)
t.forward(50)
t.end_fill()
```
